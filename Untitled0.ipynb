```python
import numpy as np
num_samples = 1000000
inside_count = 0
np.random.seed(13)
array_x = np.random.uniform(-1, 1, num_samples)
array_y = np.random.uniform(-1, 1, num_samples)
for i in range(num_samples):
    if array_x[i]**2 + array_y[i]**2 <= 1:
        inside_count += 1
est_pi = 4 * inside_count / num_samples
print("推定された円周率の値：", est_pi)
```

```python
import numpy as np
import statsmodels.stats.api as sm
mu0, sigma = 60, 15
n = 25
m = 10000
np.random.seed(13)
Count = 0
for trial in range(m):
    array_x = np.random.normal(mu0, sigma, n)
    dsw = sm.DescrStatsW(array_x)
    t, p, _ = dsw.ttest_mean(value=mu0)
    if p <= 0.05:
        Count += 1
print("結果：",Count / m)
```

```python
import statsmodels.api as sm
import pandas as pd
df = pd.read_csv('data7-3.csv')
df_X = df[['Media_A', 'Media_B', 'Media_C', 'Price']]
df_y = df['Sales']
df_X = sm.add_constant(df_X)
model = sm.OLS(df_y, df_X)
results = model.fit()
display(results.summary())
```

```python
import numpy as np
import statsmodels.stats.api as sm
mu, sigma = 60, 15
n = 3000
m = 50000
np.random.seed(13)
true_count = 0
for trial in range(m):
    array_x = np.random.normal(mu, sigma, n)
    dsw = sm.DescrStatsW(array_x)
    interval = dsw.tconfint_mean(alpha=0.05)
    if interval[0] <= mu <= interval[1]:
        true_count += 1
print("結果：", true_count /m)
```
